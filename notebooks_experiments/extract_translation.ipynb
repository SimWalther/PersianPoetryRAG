```python
from docx import Document
from glob import glob
from docx.enum.text import WD_ALIGN_PARAGRAPH
import pandas as pd

files = glob("../english/book*/*.docx")

def parse_docx(file_path):
    doc = Document(file_path)
    book_list = []
    line_no_list = []
    is_title = True
    is_first_beyt = True
    is_second_beyt = False
    first_beyt_list = []
    second_beyt_list = []
    count = 0

    for para in doc.paragraphs:
        count += 1
        text = para.text

        if "Rumi, Mathnavi" in text:
            is_title = False
            is_first_beyt = True
            book, line = text.split("Book ")[-1].split(", Line ")
            # print(f"Book {book}, line {line}")
            book_list.append(int(book))
            line_no_list.append(int(line))
        elif para.alignment == WD_ALIGN_PARAGRAPH.CENTER:
            is_title = True
            #print("TITLE: ", text)
        elif len(para.text) == 0 or para.text.isspace():
            is_title = True
        else:
            if is_title:
                #print("TITLE: ", text)
                pass
            else:
                if is_first_beyt:
                    #print(f"x", end="")
                    first_beyt_list.append(text)
                    is_first_beyt = False
                    is_second_beyt = True
                elif is_second_beyt:
                    if count == 6917:
                        print(text)
                    #print(f"_{count}", end="")
                    second_beyt_list.append(text)

        # print(len(first_beyt_list) - len(second_beyt_list))
    
    return {
        'book': book_list,
        'line': line_no_list,
        'beyt1_en': first_beyt_list,
        'beyt2_en': second_beyt_list
    }

result_list = []

for doc in ["../english/book 2/1901-3810_Book2_EN.docx"]:
    result = parse_docx(doc)

    print(doc)
    for k,v in result.items():
        print(k, len(v))
        
    result_df = pd.DataFrame(result)
    result_list.append(result_df)

df = pd.concat(result_list, axis=0)
df
```

```python
import pandas as pd
import psycopg2

db_host = "localhost"
db_name = "PersianPoetry"
db_user = "simon"
db_password = "shehr farsi"

try:
    # Connect to PostgreSQL
    conn = psycopg2.connect(host=db_host, database=db_name, user=db_user, password=db_password)
    cursor = conn.cursor()

    # Prepare the data for insertion
    # First, let's make sure we have the right columns and format
    update_data = merged_df[['book', 'number', 'beyt1_en', 'beyt2_en']].copy()
    
    # Convert to list of tuples for efficient bulk update
    records = [tuple(row) for row in update_data.values]
    
    # Create the UPDATE query
    update_query = """
        UPDATE masnavi 
        SET beyt1_en = %s, beyt2_en = %s
        WHERE book = %s AND number = %s
        
    """
    
    # Execute the update for each record
    for record in records:
        cursor.execute(update_query, record[2:]+ record[:2])
    
    # Commit the changes
    conn.commit()
    print(f"Successfully updated {len(records)} records in masnavi table")

    cursor.close()
    conn.close()

except psycopg2.Error as e:
    print(f"Error connecting to PostgreSQL: {e}")
```
