```python
from datasets import load_dataset
dataset = load_dataset("mabidan/ganjoor")
```

```python
import re

ZERO_WIDTH_SPACE = u"\u200c"

tokens = []

for text in dataset["train"]["text"]:
    if not text:
        continue

    text = text.replace(ZERO_WIDTH_SPACE, " ")
    tokens.append(re.split("\\W", text))
```

```python
import numpy as np
tokens = np.concatenate(tokens, axis=None)
```

```python
from collections import Counter
tokens_count = Counter(tokens)
```

```python
import matplotlib.pyplot as plt
import arabic_reshaper
from bidi import algorithm as bidialg

top_50 = tokens_count.most_common(50)
top_50_tokens = [bidialg.get_display(arabic_reshaper.reshape(t[0])) for t in top_50]
top_50_count = [t[1] for t in top_50]

plt.figure(figsize=(14, 4), layout='constrained')
plt.bar(top_50_tokens, top_50_count)
plt.yscale("log")
plt.xticks(np.arange(0, 50, 1));
```
