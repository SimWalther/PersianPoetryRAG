```python
from docx import Document
import re

def process_document(file_path):
    # Load the document
    doc = Document(file_path)
    
    # Track the current line number to increment
    current_line_num = None
    
    # Process each paragraph in the document
    for para in doc.paragraphs:
        text = para.text
        
        # Check if paragraph starts with "Rumi, Mathnavi, Book 4, Line"
        if text.startswith("Rumi, Mathnavi, Book 4, Line"):
            # Extract the current line number
            match = re.search(r'Line (\d+)', text)
            if match:
                line_num = int(match.group(1))
                
                # If this is the first occurrence, set initial line number
                if current_line_num is None:
                    current_line_num = line_num
                
                # Replace with incremented line number
                new_text = re.sub(
                    r'Line \d+',
                    f"Line {current_line_num}",
                    text
                )
                
                # Update the paragraph text
                para.text = new_text
                
                # Increment for next occurrence
                current_line_num += 1

    # Save the modified document
    doc.save(file_path)

# Example usage:
process_document("../english/book 4/1951-3855.docx")
```
