## Frequent words

```python
from shekar import Normalizer, WordTokenizer, Lemmatizer
import numpy as np
from tqdm import tqdm
from collections import Counter

from shekar.preprocessing import (
  PunctuationRemover,
  NonPersianRemover,
  DiacriticRemover,
)

clean_text = NonPersianRemover() | DiacriticRemover() | PunctuationRemover()

normalizer = Normalizer()
tokenizer = WordTokenizer()
lemmatizer = Lemmatizer()

def preprocess_persian(text):
    text = clean_text(text)
    text = normalizer(text)
    tokens = tokenizer(text)
    tokens = [lemmatizer(token) for token in list(tokens)]
    return tokens
```

```python
all_tokens = []

for ghazal_idx, ghazal_row in tqdm(ghazal.iterrows()):
    all_tokens += preprocess_persian((str(ghazal_row.beyt1 + ghazal_row.beyt2)))

token_count = Counter(all_tokens)
```

```python
import arabic_reshaper
from bidi import algorithm as bidialg

most_common_tokens = token_count.most_common(50)
plt.figure(figsize=(10, 3))
plt.title("Ghazal most common tokens")
plt.bar(
    [bidialg.get_display(arabic_reshaper.reshape(token)) for token, _ in most_common_tokens],
    [count for _, count in most_common_tokens],
    color="#aca"
)
plt.xlabel("Token")
plt.ylabel("Number of occurences")
plt.xticks(rotation=90);
```
