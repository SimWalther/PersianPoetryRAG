```python
from langchain_huggingface import HuggingFaceEmbeddings

embeddings = HuggingFaceEmbeddings(model_name="Qwen/Qwen3-Embedding-0.6B")
```

```python
masnavi.query("beyt2.isna()")
```

```python
masnavi = masnavi.dropna(axis=0)
```

```python
import numpy as np
from numpy.linalg import norm
from tqdm import tqdm

def cosine_similarity(a, b):
    return (a @ b.T) / (norm(a)*norm(b))

results = []
for masnavi_idx, masnavi_row in masnavi.iterrows():
    beyt1_cos_sim = cosine_similarity(
        np.asarray(embeddings.embed_query(masnavi_row.beyt1)), 
        np.asarray(embeddings.embed_query(masnavi_row.beyt1_en)),
    )
    beyt2_cos_sim = cosine_similarity(
        np.asarray(embeddings.embed_query(masnavi_row.beyt2)), 
        np.asarray(embeddings.embed_query(masnavi_row.beyt2_en)),
    )
    
    results.append({
        'index': masnavi_idx,
        'beyt1_similarity': beyt1_cos_sim,
        'beyt2_similarity': beyt2_cos_sim
    })

# Assign all at once
for result in results:
    masnavi.loc[result['index'], 'beyt1_similarity'] = result['beyt1_similarity']
    masnavi.loc[result['index'], 'beyt2_similarity'] = result['beyt2_similarity']
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(3, 4))
plt.title("Beyts-Translation Cosine Similarity")
plt.boxplot(masnavi.beyt1_similarity + masnavi.beyt2_similarity)
plt.xlabel('Beyt')
plt.xticks([])
```

```python
import pandas as pd

masnavi = pd.read_parquet('masnavi_with_similarity.parquet')
```

```python
import matplotlib.pyplot as plt

for book in masnavi.book.unique():
    book_sim_1 = masnavi.query('book == @book').set_index('number').sort_index().beyt1_similarity
    book_sim_2 = masnavi.query('book == @book').set_index('number').sort_index().beyt2_similarity

    plt.title(f'Book {book} Beyt-Translation Similarity')
    plt.plot(book_sim_1, alpha=0.7, label='Beyt1')
    plt.plot(book_sim_2, alpha=0.7, label='Beyt2')
    plt.plot(((book_sim_1 + book_sim_2) / 2).rolling(50).mean(), color='black', label='Rolling Mean (50)')
    plt.xlabel("Line Number")
    plt.ylabel("Cosine Similarity")
    plt.grid()
    plt.legend()
    plt.show()
    plt.close()
```
