```python
from shekar import Normalizer, WordTokenizer, Stemmer
from tqdm import tqdm

normalizer = Normalizer()
stemmer = Stemmer()
tokenizer = WordTokenizer()

all_tokens = []

for masnavi_idx, masnavi_row in tqdm(masnavi.iterrows()):
    beyt = masnavi_row.beyt1
    beyt2 = masnavi_row.beyt2

    if beyt2:
        beyt += masnavi_row.beyt2
    
    normalized_text = normalizer(str(beyt))
    tokens = list(tokenizer(normalized_text))
    tokens = [stemmer(token) for token in tokens]

    all_tokens += tokens
```

```python
import numpy as np
all_tokens = np.unique(all_tokens)
```

```python
from shekar.embeddings import WordEmbedder
from numpy.linalg import norm
def cosine_similarty(a, b):
    return (a @ b.T) / (norm(a)*norm(b))

embedder = WordEmbedder(model="fasttext-d300")

first_word = "عشق"
second_word = "محبت"
third_word = "زندگی"

random_word = np.random.choice(all_tokens, 1)[0]
embedding_first_word = embedder(first_word)
embedding_second_word = embedder(second_word)
embedding_third_word = embedder(second_word)
embedding_random = embedder(random_word)

print(f"Cos. Similarity {first_word} <=> {second_word}: {cosine_similarty(embedding_first_word, embedding_second_word)}")
print(f"Cos. Similarity {first_word} <=> {third_word}: {cosine_similarty(embedding_first_word, embedding_third_word)}")
print(f"Cos. Similarity {first_word} <=> {random_word}: {cosine_similarty(embedding_first_word, embedding_random)}")
```

```python
all_tokens_with_emb = []
embeddings = []

for token in all_tokens:
    embedding = embedder(token)

    if np.any(embedding):
        all_tokens_with_emb.append(token)
        embeddings.append(embedding)
```

```python
embeddings = np.stack(embeddings)
```

```python
import umap

reducer = umap.UMAP(n_neighbors=30)

embeddings_umap = reducer.fit_transform(embeddings)
```

```python
min_x, min_y = embeddings_umap.min(axis=0)
max_x, max_y = embeddings_umap.max(axis=0)
```

```python
import matplotlib.pyplot as plt
import arabic_reshaper
from bidi import algorithm as bidialg

def find_token(word):
    for i, token in enumerate(all_tokens_with_emb):
        if token == word: 
            return i
    
    return -1

plt.title("2d UMAP Projection of Masnavi Tokens")
plt.scatter(embeddings_umap[:, 0], embeddings_umap[:, 1], c="#aca")

for word in ["سمندر", "آتش", "اسب", "مهر", "محبت", "عشق", "مرد", "سیاه", "سفید", "گوسفند", "شکر"]:
    emb = embeddings_umap[find_token(word)]
    plt.text(emb[0], emb[1], s=bidialg.get_display(arabic_reshaper.reshape(word)))

plt.xticks(np.linspace(min_x, max_x, 10));
plt.yticks(np.linspace(min_y, max_y, 10));
```
